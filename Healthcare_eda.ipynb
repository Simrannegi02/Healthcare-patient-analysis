```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("/content/patient1.xlsx",skiprows=3)
df.head()
```

```python
df['Admission_Date']=pd.to_datetime(df['Admission_Date'])
df['Discharge_Date']=pd.to_datetime(df['Discharge_Date'])
df['Stay Days']=(df['Discharge_Date']-df['Admission_Date']).dt.days
df[['Admission_Date','Discharge_Date','Stay Days']].head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Disease'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
plt.figure(figsize=(6,4))
df['Age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(False)
plt.show()

```

```python
df['Gender'].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Patients")
plt.grid(False)
plt.show()
```

```python
df["Disease"].value_counts().plot(kind="barh")
plt.title("Disease Frequency")
plt.xlabel("Disease")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```

```python
df["Month"]=df["Admission_Date"].dt.to_period("M").astype(str)
df["Month"].value_counts().sort_index().plot(kind="line")
plt.title("Monthly Admission Trend")
plt.xlabel("Month")
plt.ylabel("Patients")
plt.grid(False)
plt.show()
```

```python
print("Most common Disease:", df["Disease"].mode()[0])
print("Average Age:", df["Age"].mean())
print("Average Hospital stay :", df["Stay Days"].mean())
print("Longest Stay:", df["Stay Days"].max())
print("Shortest Stay:", df["Stay Days"].min())
print("Gender Distribution:\n", df["Gender"].value_counts())
print("City-wise Patients:\n", df["City"].value_counts())
```
